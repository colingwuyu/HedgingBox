# Market Impact

## Price Dynamic
1. Agent's action $A_{t-1}$ based on stock price $S_{t-1}$.  
2. Market impact caused by $A_{t-1}$:  
$$\Delta{P_{t-1}}=sign(A_{t-1})\alpha\sigma\sqrt{{\lvert}\frac{A_{t-1}}{V}{\rvert}}$$
3. Stock price after executing $A_{t-1}$:
$$S^{'}_{t-1}=S_{t-1}exp(\Delta{P_{t-1}})$$
4. Use GBM to simulate next step $S_t$ based on $S^{'}_{t-1}$:
$$S_t=S^{'}_{t-1}exp((\mu-\frac{\sigma^2}{2})\Delta{t}+\sigma W_{\Delta{t}})$$

Stock price is generated by a vanilla GBM dynamic $G_t$ combining with market impacts as:
$$S_t=G_texp(\sum_{i<t}\Delta{P_i})$$

## P&L
1. Daily P&L reward is calculated as:  

$$PnL_t = V_t - V_{t-1} + H_t (S_t - S_{t-1}) - TC_t + I_t$$

  
 + $V_t$: derivative price at time t
 + $S_t$: stock price at time t (see Market impact)
 + $H_t$: stock holding at time t. $H_t=H_{t-1}+A_{t-1}$
 + $TC_t$: transaction cost $TC_t=\alpha S_{t-1} A_{t-1}$
 + $I_t$: interests accrued from cash account $I_t=B^{+}_{t-1} \times \delta r_t$


2. Cash account balance is accumulated as:
$$B^{+}_{t-1}=B^{-}_t+CF_{t-1}$$
  
    where cash flow $CF_{t-1}$ includes cash proceeds and transaction cost of $A_{t-1}$:
    
$$CF_{t-1}=-S^{'}_{t-1} \times A_{t-1} -TC_t$$

3. At last step T, calculate P&L for the market impact caused by dumping portfolio:

$$PnL^{*}_T = H_T \times (S^{'}_T - S_T)$$

4. Accumulated P&L for the episode:

$$PnL = \sum_{t=0}^{T}{PnL_t}+PnL^{*}_T$$
